## **1. Giới thiệu EDA và mục tiêu phân tích**

- **Mục đích:** Nêu vai trò của EDA trong Lab 02 và đầu ra phục vụ Tableau.
- **Nội dung nên làm:** Giới thiệu ngắn về bộ dữ liệu WDI, phạm vi phân tích, và câu hỏi chính.
- **Output mong đợi:** Định hướng rõ cho toàn notebook.

## **2. Nguồn dữ liệu và file đầu vào**

- **Mục đích:** Xác định rõ dữ liệu nào được dùng cho EDA.
- **Nội dung nên làm:** Giới thiệu `wdi_long_clean.csv` và `wdi_wide_for_tableau.csv`.
- **Output mong đợi:** Mô tả ngắn về nguồn dữ liệu và cách dùng từng file.

## **3. Đọc dữ liệu đã tiền xử lý**

- **Mục đích:** Kiểm tra dữ liệu sạch đã sẵn sàng cho phân tích chưa.
- **Nội dung nên làm:** Đọc file, xem số dòng/cột, kiểu dữ liệu, vài dòng mẫu.
- **Output mong đợi:** Tóm tắt nhanh cấu trúc dữ liệu sau tiền xử lý.

## **4. Tổng quan cấu trúc bộ dữ liệu**

- **Mục đích:** Nắm bức tranh toàn cục của dữ liệu.
- **Nội dung nên làm:** Số quốc gia, số năm, số chỉ số, số region, số income group.
- **Output mong đợi:** Bảng/tóm tắt tổng quan dataset.

## **5. Từ điển chỉ số và nhóm biến**

- **Mục đích:** Giải thích ý nghĩa các chỉ số để đọc EDA đúng ngữ cảnh.
- **Nội dung nên làm:** Liệt kê nhóm chỉ số chính, tên ngắn, đơn vị, ý nghĩa phân tích.
- **Output mong đợi:** Bảng dictionary chỉ số hỗ trợ đọc biểu đồ và dashboard.

## **6. Kiểm tra phạm vi và độ phủ dữ liệu**

- **Mục đích:** Xác nhận dữ liệu có đủ theo chiều năm, quốc gia, region, income group.
- **Nội dung nên làm:** Kiểm tra năm min/max, số quốc gia theo khu vực, số nhóm thu nhập.
- **Output mong đợi:** Bảng hoặc biểu đồ cho thấy mức độ phủ dữ liệu.

## **7. Kiểm tra dữ liệu thiếu và chất lượng dữ liệu**

- **Mục đích:** Đánh giá độ tin cậy của dữ liệu trước khi trực quan hóa.
- **Nội dung nên làm:** Tỷ lệ missing theo chỉ số, theo năm, theo quốc gia, giá trị bất thường.
- **Output mong đợi:** Danh sách vấn đề dữ liệu cần lưu ý khi vẽ dashboard.

## **8. Thống kê mô tả theo từng chỉ số**

- **Mục đích:** Hiểu phân phối của từng indicator.
- **Nội dung nên làm:** Min, max, mean, median, độ lệch, phân vị cho các chỉ số chính.
- **Output mong đợi:** Bảng thống kê mô tả và nhận xét ngắn.

## **9. Phân tích theo quốc gia, khu vực và nhóm thu nhập**

- **Mục đích:** So sánh sự khác biệt giữa các nhóm đối tượng.
- **Nội dung nên làm:** Xếp hạng quốc gia, so sánh region, so sánh income group.
- **Output mong đợi:** Bảng/biểu đồ so sánh để kể chuyện trên dashboard.

## **10. Phân tích xu hướng theo thời gian**

- **Mục đích:** Xem biến động của các chỉ số qua các năm.
- **Nội dung nên làm:** Trend line theo quốc gia, region, income group; nhận diện tăng/giảm dài hạn.
- **Output mong đợi:** Nhận xét về xu hướng và giai đoạn nổi bật.

## **11. Phân tích mối quan hệ giữa các chỉ số**

- **Mục đích:** Tìm các cặp chỉ số đáng chú ý để dùng scatter/correlation trong Tableau.
- **Nội dung nên làm:** Tương quan giữa kinh tế, y tế, giáo dục, môi trường.
- **Output mong đợi:** Danh sách quan hệ mạnh/yếu và gợi ý biểu đồ phù hợp.

## **12. Gợi ý trường dữ liệu cho Tableau**

- **Mục đích:** Chuyển kết quả EDA thành cấu trúc dùng được cho dashboard.
- **Nội dung nên làm:** Gợi ý field làm dimension, measure, filter, parameter.
- **Output mong đợi:** Checklist field cho Tableau và cách dùng chúng.

## **13. Tổng kết EDA và lưu ý khi trực quan hóa**

- **Mục đích:** Chốt lại insight quan trọng và cảnh báo khi làm dashboard.
- **Nội dung nên làm:** Tóm tắt phát hiện chính, giới hạn dữ liệu, lưu ý về missing values và đơn vị.
- **Output mong đợi:** Phần kết luận ngắn, định hướng sang bước dựng dashboard.
